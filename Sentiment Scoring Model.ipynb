```python
##store tweet text
#this step uses modified code taken from https://towardsdatascience.com/step-by-step-twitter-sentiment-analysis-in-python-d6f650ade58d
tweetList=[]
posList=[]
negList=[]
neuList=[]
largestSent=0
tweetSample=df["content"].loc[df["id"]<=500]
for tweet in tweetSample:
    tweetList.append(tweet)
    score=SentimentIntensityAnalyzer().polarity_scores(tweet)
    neg=score['neg']
    neu=score['neu']
    pos=score['pos']
    if pos > neg:
        posList.append(tweet)
    elif neg > pos:
        negList.append(tweet)
    elif pos == neg:
        neuList.append(tweet)
##output sentiment percentages 
percentPos= ((len(posList)/len(tweetList))*100)
percentNeg= ((len(negList)/len(tweetList))*100)
percentNeu= ((len(neuList)/len(tweetList))*100)
#Code borrowed from https://towardsdatascience.com/step-by-step-twitter-sentiment-analysis-in-python-d6f650ade58d ends here
##print sentiment percentages and establish a score for the largest sentiment percentage
print(f"positive sentiment: {round(percentPos,2)}%")
print(f"negative sentiment: {round(percentNeg,2)}%")
print(f"neutral sentiment: {round(percentNeu,2)}%")
if percentPos>percentNeg and percentPos>percentNeu:
    print ("Overall sentiment is positive")
    largestSent=3
elif percentNeg>percentPos and percentNeg>percentNeu:
    print("Overall sentiment is negative")
    largestSent=1
elif percentNeu>percentPos and percentNeu>percentNeg:
    print("Overall sentiment is neutral")
    largestSent=2
else:
    print("Overall sentiment is mixed")
```
