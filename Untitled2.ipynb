```python
import pandas as pd
import numpy as np
```

```python
csv=pd.read_csv("stockdatacleaned.csv")
csv
```

```python
df=csv[["Symbol","Date","Open","Close","High","Low"]]
df=df.rename(columns={'Open':'Opening_Price_USD',"Close":"Closing_Price_USD",'High':'High_USD',"Low":"Low_USD"})
df=df.insert(0,'indexkey',df.index)
df.to_csv("StockPricesUSD.csv",index=False)
```

```python
df1=csv[["Symbol","Date","Volume"]]
df1=df1.rename(columns={"Volume":"Volume_Traded"})
df1["indexkey"]=df.index
df1.to_csv("StockVolume.csv",index=False)
```
